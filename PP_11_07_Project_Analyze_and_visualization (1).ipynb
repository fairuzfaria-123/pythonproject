```python
#ans no 1:
#Number of students who have completed both ssc and hsc with gpa greater than 4

df=pd.read_csv('student_data.csv')
completed_both = df[(df['ssc_gpa'] > 4) & (df['hsc_gpa'] > 4)]
num_students=len(completed_both)
print(f"Number of students who have completed both SSC and HSC with GPA greater than 4:", num_students)

```

```python
#ans no 2:
#display all students who have attended study abroad in Canada
df=pd.read_csv('student_data.csv')
canada_students=df [df['study_abroad_country'] == 'Canada']
canada_students_info=canada_students[['id', 'first_name', 'last_name']]

print(canada_students_info)
```

```python
#ans no 3:
#average CGPA of Students from electrical engineering
df= pd.read_csv('student_data.csv')
electrical_engineering = df [df['department'] == 'Electrical Engineering']
average_cgpa = df["cgpa"].mean()
print(f"Average cgpa of electric engineering students: {average_cgpa}")
```

```python
#ans no 4:
#replace missing value in the "study_abroad_city" column with "unknown"
df = pd.read_csv("student_data.csv")
df.isnull().sum()
print(df["study_abroad_city"])
df_filled =df.fillna({"study_abroad_city": "Unknown"})
df_filled.isnull().sum()

```

```python
#ans no 5:
#pie chart showing the distribution of students by graduation_date
import matplotlib.pyplot as plt
graduation_counts = df['graduation_date'].value_counts()
plt.figure(figsize=(8,8 ))
plt.pie(graduation_counts, labels=graduation_counts.index, autopct='%1.0f%%', startangle=180)
plt.title('Distribution of Students by Graduation Date')
plt.show()
```
